```python
import cv2
```

```python
img = cv2.imread('wallpaper.jpg')
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.resize(img,(860,620))
img.shape
```

**gray has only one channel of range (0,255)** where 0 = black and 255 = white
**BRG has 3 channels(**Blue,Red,Green) of color ranging (0,255).
Due to complexity of BRG we use gray because it reduce img to single channel but the problem is we will lose color, **so to change it into single channel and still have color we use hsv formate of image.**

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(gray)
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv2_imshow(hsv)
```
