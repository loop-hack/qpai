```python
# Python Lists

mylist = ["Ankit", "Anurag", "is", "my", "name", "."]

print(mylist)

print(len(mylist))

print(mylist[0])
print(mylist[0::2])
print(mylist[-1::-1])
```

```python
# Python - Change List Items

"""Change some specific value"""

mylist[0] = "Vipin"
print(mylist)

"""Change a range of items"""

mylist[0:2] = ["Vipin","Ankit"]
print(mylist)

mylist.insert(0, "Vipin")
print(mylist)
```

```python
# Python add items 

a = ["1", "2", "3", "4", "5"]

a.append("6")
print(a)

a.insert(0,"0")
print(a)

b = ["7", "8", "9", "10"]
a.extend(b)
print(a)

# removing from list

a.remove("0")
print(a)

a.pop(9)
print(a)

del a[8]
print(a)

a.clear()
print(a)
```

```python
# Accesing List using loop

a = ["a", "b", "c", "d", "e"]

for i in a:
    print(i)

for n in range(len(a)):
    print(a[n])


# Using while loop

i=0
while i<len(a):
    print(a[i])
    i += 1
```
