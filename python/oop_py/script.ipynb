```python
class Dog:
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed

    def bark(self):
        print("Whoof Whoof")

dog1 = Dog("Galaxy_Destroyer" ,"husky")  # we have created an object naming dog1 from the class Dog()
dog1.bark()
print(dog1.name)
print(dog1.breed)

dog2 = Dog("Shivam", "Labrador")
dog2.bark()
print(dog2.name)
print(dog2.breed)
```
