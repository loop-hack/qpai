```python
class Dog:
    def __init__(self, name, breed, owner):
        self.name = name
        self.breed = breed
        self.owner = owner

    def bark(self):
        print("Whoof Whoof")

class Owner:
    def __init__(self, name,address, contact_number):
        self.name = name
        self.address = address
        self.number = contact_number

owner1 = Owner("Danny", "122 Springfield Drive", "888-999")
dog1 = Dog("Galaxy_Destroyer" ,"husky", owner1)  # we have created an object naming dog1 from the class Dog()
print(dog1.owner.name)

owner2 = Owner("Anup", "Boys Hostel, SEC Sasaram", "823-999")
dog2 = Dog("Shivam", "Labrador", owner2)
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        print(f"Hello my name is {self.name} and I'm {self.age} years old.")

person1 = Person("Alice", 31)
person1.greet()
```

```python
class User:
    def __init__(self, username, email, password):
        self.username = username
        self.email = email
        self.password = password

    def say_hi_to_user(self,user):
        print(f"Sending massage to {user.username}: Hi {user.username} I'm {self.username}")

user1 = User("Batman", "timetraveller1732@gmail.com", "Subject")
user2 = User("Anup", "anup33333@gmail.com", "888888")

user1.say_hi_to_user(user2)

user1.email="Ankit"  # we can change data of an object in class
print(user1.email)
```

```python
# we use _ in python to make data protected:

class User:
    def __init__(self, username, email, password):
        self.username = username
        self._email = email     # _ before email protect data stored in email
        self.password = password

    def clean_email(self):
        return self._email.lower().strip()

user1 = User("Batman", "TiMetrAvellEr1732@gmail.com", "1234")
print(user1._email)      # we are able do this even it's protected, it's a problem
print(user1.clean_email())
```

```python
class Student:
    def __init__(self, name, registration, email, password):
        self.name = name
        self.registration = registration
        self.__email = email #__ means it is private _ means it is protected
        self.__password = password
    
    def clean_email(self):
        return self.__email.lower().strip()
    
s1 = Student("Ankit", 24105124027, "TiMetrAvellEr1732@gmail.com", "123344")
print(s1.name, s1.registration)
""" protected (_): it still could be accessible outside of class _email = give email
    private(__): This can't be protected outside of the class at all __email = ERROR"""
```
