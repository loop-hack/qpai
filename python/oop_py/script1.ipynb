```python
class User:
    def __init__(self, username, email, password):
        self.username = username 
        self._email = email
        self.password = password

    @property
    def email(self):   # @property followed by method is a Python way of creating getter
        print("Email accessed")
        return self._email
    
    @email.setter      #setter
    def email(self,new_email):
        if "@" in new_email:
            self._email = new_email
    
user1 = User("Batman","timetraveller1732@gmail.com", "18829")
user1.email = "hey"         #This will not change email bcz it doen't have @ symbol init
user1.email = "Batman@gmail.com"     #This will change the email.
print(user1.email)
```

Static Attributes:

```python
""" Definition : A Static attribute (sometimes called a class attribute) is a attribute
that belongs to the class itself, not to any specific instance of the class."""

class User:
    user_count = 0

    def __init__(self, username, email):
        self.username = username
        self.email = email
        User.user_count += 1

    def display_user(self):
        print(f"Username: {self.username}, Email: {self.email}")

user1 = User("Batman", "timetraveller1732@gmail.com")
user2 = User("Superman", "ankit@gmail.com")
user1.display_user()
user2.display_user()


print(User.user_count)  # we can access static attribute created using class
print(user1.user_count) # we can also access static attribute using instances
```

```python
# Static vs Instance method Example

class BankAccount:
    MIN_BALANCE = 100

    def __init__(self, owner, balance = 0):
        self.owner = owner
        self._balance = balance

    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            print(f"{self.owner}'s new balance ${self._balance}")
        else:
            print("deposit must be positive.")

    @staticmethod  # This is a way of creating static instances
    def is_valid_interest_rate(rate):
        return 0<= rate <=5
    
account = BankAccount("Ankit", 10000000)
account.deposit(478798)

print(BankAccount.is_valid_interest_rate(3))
print(BankAccount.is_valid_interest_rate(2.65))
```
